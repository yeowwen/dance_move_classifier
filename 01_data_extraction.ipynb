### Create User Defined Function: process_video and update_data
---

* For data extraction and video processing

* Extract the landmarks data [0, 9, 10, 12, 14, 16, 11, 13, 15, 24, 26, 28, 23, 25, 27] for x,y,z coordinates and visibility


**update_data**

Parameters:

* video_path
* move_name
* video_num
* bboy_name
* frame_num_start
* frame_num_end
* dataset_numpy_path

Function:

1. Constructs paths for the video, CSV file, and landmark videos.

2. If the full_video_path exist, then call out user defined function process_video to process the specified video, else prints a message if the video file is not found.

3. Adds a new row to the moves_df with the paths and details of the processed video.

4. The updated moves_df is saved as a CSV file.

5. The function also measures the execution time.


**process_video**

Parameters:

* video_path
* frame_num_start
* frame_num_end
* video_num
* move_name
* bboy_name
* dataset_numpy_path
* dataset_csv_path
* videolandmarks_path

Function:

1. Read the video file from video_path.

2. Make directories to store the output data if they don't already exist.

3. Retrieving the video frames from frame_num_start to frame_num_end.

4. Use MediaPipe to detect human poses in each selected frame.

5. If landmarks are detected, save them as NumPy files and prepare data for the CSV file. If no landmarks are detected, fill with zeros.

6. Draw the detected landmarks on the video frames and write these annotated frames to a new video file. Also, include text overlay such as move_name, video_num, bboy_name and frame number.

7. Compile all the landmark data into a CSV file and save it.

8. Close the video file and release resources used for video processing.

```python
# process_video function
def process_video(video_path, frame_num_start, frame_num_end, video_num, move_name, bboy_name, dataset_numpy_path, dataset_csv_path, videolandmarks_path):
    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    # Create directories
    numpy_save_path = os.path.join(dataset_numpy_path, move_name, str(video_num))
    csv_save_path = os.path.join(dataset_csv_path, move_name)
    landmarks_video_path = os.path.join(videolandmarks_path, move_name)
    os.makedirs(numpy_save_path, exist_ok=True)
    os.makedirs(csv_save_path, exist_ok=True)
    os.makedirs(landmarks_video_path, exist_ok=True)

    # MediaPipe setup
    mp_pose = mp.solutions.pose
    pose = mp_pose.Pose()
    # Drawing specifications
    mp_drawing = mp.solutions.drawing_utils
    landmark_drawing_spec = mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2)
    connection_drawing_spec = mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2)


    # Prepare CSV data and video writer
    csv_data = []
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out_video = cv2.VideoWriter(os.path.join(landmarks_video_path, f"{move_name}_{video_num}_landmark.mp4"), fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))
    
    # creates a loop that keeps running as long as the video file is open and readable
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if frame_num_start <= frame_count <= frame_num_end:
            # Process frame for landmarks
            results = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

            # Initialize default landmarks (all zeros)
            default_landmarks = np.zeros((15, 4))  # 15 landmarks, each with x, y, z, visibility

            # Check if landmarks are detected
            if results.pose_landmarks:
                landmarks = results.pose_landmarks.landmark
                extracted_landmarks = np.array([(landmark.x, landmark.y, landmark.z, landmark.visibility) for landmark in landmarks])[[0, 9, 10, 12, 14, 16, 11, 13, 15, 24, 26, 28, 23, 25, 27]]
            else:
                extracted_landmarks = default_landmarks

            # Flatten and save landmarks
            flattened_landmarks = extracted_landmarks.flatten()
            np.save(os.path.join(numpy_save_path, f"{frame_count}.npy"), flattened_landmarks)

            # Add data to CSV
            csv_row = [bboy_name, move_name, frame_count] + flattened_landmarks.tolist()
            csv_data.append(csv_row)
            
            
            #cv2.putText(frame, f"Frame: {frame_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
            text_y_offset = 70
            line_spacing = 80
            cv2.putText(frame, f"Move: {move_name}", (10, text_y_offset), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 255, 255), 3, cv2.LINE_AA)
            cv2.putText(frame, f"Video Num: {video_num}", (10, text_y_offset + line_spacing), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 255, 255), 3, cv2.LINE_AA)
            cv2.putText(frame, f"Bboy Name: {bboy_name}", (10, text_y_offset + 2 * line_spacing), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 255, 255), 3, cv2.LINE_AA)
            cv2.putText(frame, f"Frame: {frame_count}", (10, text_y_offset + 3 * line_spacing), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 255, 255), 3, cv2.LINE_AA)

            
            
            # Draw landmarks and save video frames   
            if results.pose_landmarks:
                mp_drawing.draw_landmarks(
                    frame,
                    results.pose_landmarks,
                    mp_pose.POSE_CONNECTIONS,
                    landmark_drawing_spec,
                    connection_drawing_spec
                )
            out_video.write(frame)
            
        # Prepare for the next frame
        frame_count += 1

    cap.release()
    out_video.release()
    
    # Save CSV data
    columns = ['bboy_name', 'move_name', 'frame'] + [f'p{idx}{coord}' for idx in [0, 9, 10, 12, 14, 16, 11, 13, 15, 24, 26, 28, 23, 25, 27] for coord in ['x', 'y', 'z', 'v']]
    df = pd.DataFrame(csv_data, columns=columns)
    df.to_csv(os.path.join(csv_save_path, f"{move_name}_{video_num}.csv"), index=False)

    print(f"Processed video {video_path}. Data saved to {numpy_save_path}, {csv_save_path}, and {landmarks_video_path}")

    
# update_data function
def update_data(video_name, move_name, video_num, bboy_name, frame_num_start, frame_num_end, moves_df):
    # Start measure execution time
    start_time = time.time()
    
    # Construct the relative paths
    relative_video_path = os.path.join("archive_video", move_name, video_name + '.mp4')
    relative_csv_path = os.path.join("dataset_csv", move_name, f"{video_name}.csv")
    relative_videolandmarks_path = os.path.join("video_landmarks", move_name, f"{move_name}_{video_num}_landmark.mp4")
    
    # Full paths for processing the video
    full_video_path = os.path.join(video_path, move_name, video_name + '.mp4')
    full_csv_path = os.path.join(dataset_csv_path, move_name, f"{video_name}.csv")
    full_videolandmarks_path = os.path.join(videolandmarks_path, move_name)

    # If the full_video_path exists, then process the specified video
    if os.path.exists(full_video_path):
        process_video(full_video_path, frame_num_start, frame_num_end, video_num, move_name, bboy_name, dataset_numpy_path, dataset_csv_path, videolandmarks_path)
        
        # Adds a new row to the moves_df with the relative paths and details of the processed video
        new_row = {
            'videolandmarks_path': relative_videolandmarks_path,
            'videolandmarks_name': f"{move_name}_{video_num}_landmark.mp4",
            'video_path': relative_video_path, 
            'video_name': video_name, 
            'dataset_csv_path': relative_csv_path,
            'dataset_csv': f"{video_name}.csv",
            'move_name': move_name,
            'video_num': video_num, 
            'bboy_name': bboy_name, 
            'frame_num_start': frame_num_start, 
            'frame_num_end': frame_num_end
        }
        moves_df.loc[len(moves_df)] = new_row
        moves_df.to_csv(os.path.join(moves_df_path, 'moves_df.csv'), index=False)
    
    else:
        print(f"{video_name} not found.")
    
    # Stop measure execution time
    end_time = time.time()
    execution_time = end_time - start_time
    
    # Formatting execution time to hh:mm:ss
    hours, remainder = divmod(execution_time, 3600)
    minutes, seconds = divmod(remainder, 60)
    print(f"Function executed in: {int(hours):02d}:{int(minutes):02d}:{int(seconds):02d}")
```
